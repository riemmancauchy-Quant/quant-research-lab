```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

print("Environment ready")
```

```python
ticker = "SPY"

data = yf.download(ticker, start="2020-01-01", end="2025-01-01")

data.head()
```

```python
daily_volatility = data["daily_return"].std()
annual_volatility = daily_volatility * np.sqrt(252)

print(f"Daily volatility: {daily_volatility:.4%}")
print(f"Annual volatility: {annual_volatility:.2%}")
plt.figure(figsize=(10, 6))
```

```python
mean_return = data["daily_return"].mean()
volatility = data["daily_return"].std()

annual_return = mean_return * 252
annual_volatility = volatility * np.sqrt(252)

print("Mean daily return:", mean_return)
print("Daily volatility:", volatility)
print("Annual return:", annual_return)
print("Annual volatility:", annual_volatility)
```

# Initial Observations

- The dataset contains adjusted close prices for SPY, QQQ, AAPL, MSFT and BTC-USD.
- BTC-USD trades every day, while stocks and ETFs only trade on business days when the market is open, that's why when I introduced normalized_prices = prices_clean / prices_clean.iloc[0], and placed the "0", it have me problems, it was using the first day, and many of this ones, had NaN for the first day of the year. That's why needed to remove the NaN's, with the command: prices_clean = prices.dropna() creating a new variable.
- This difference in trading calendars creates missing values for equity assets on weekends and market holidays.
- Absolute price charts are not ideal for comparing assets with very different nominal price levels.
- Normalized prices provide a better view of relative performance, as they show the growth of an initial investment of 1 dollar.
- AAPL, MSFT and BTC-USD showed strong relative growth over the selected period, while BTC-USD also showed significantly higher volatility.
- SPY appears more stable than individual stocks and BTC-USD, but with lower relative growth.

# Limitations

- The data was downloaded using `yfinance`, which is useful for research and learning purposes but may not be suitable for production-grade financial systems.
- The analysis only uses adjusted close prices and does not yet include intraday data, order book data, bid-ask spreads or execution prices.
- Missing values are caused by differences in trading calendars, especially between crypto assets and equities/ETFs.
- Transaction costs, slippage, taxes and liquidity constraints are not considered at this stage.
- No trading strategy has been implemented yet; this section only covers data download, inspection and visualization.
- The selected asset universe is small and mainly focused on US equities, ETFs and Bitcoin.
- Past performance does not imply future performance.
